```python
def conv(value):
    if value == 1:
        return "yes"
    else:
        return "no"

df['SeniorCitizen'] = df['SeniorCitizen'].apply(conv)
```

```python
ax = sb.countplot(x='Churn', data = df)
ax.bar_label(ax.containers[0])
plt.title("Count Of Customers by Churn")
plt.show()
```

```python
gb = df.groupby('Churn').agg({'Churn':"count"})
gb

#print(df.columns)
```

```python
plt.figure(figsize = (3,4))
plt.pie(gb['Churn'], labels = gb.index, autopct = "%1.2f%%")
plt.title("Percentage Of Churn Customers", fontsize = 10)
plt.show()
```

```python
#from the given pie chart we can conclude that 26.54% of cusomters have churned out
#so now lets explore the reasons behind it 
```

```python
plt.figure(figsize = (5,4))
gd = sb.countplot(x = "gender", data = df, hue  = "Churn")
plt.title("Churn By Gender", fontsize = 10)
gd.bar_label(gd.containers[0])
plt.show()
```

```python
plt.figure(figsize = (5,4))
gd = sb.countplot(x = "SeniorCitizen", data = df, hue  = "Churn")
plt.title("Churn By SeniorCitizen", fontsize = 10)
gd.bar_label(gd.containers[0])
plt.show()
```

```python
# Calculate total counts per group for accurate percentages
total = df['SeniorCitizen'].value_counts()

# Create a normalized (percentage) stacked bar chart
plt.figure(figsize=(4, 3))
gd = sb.histplot(data=df, x='SeniorCitizen', hue='Churn', multiple='fill', shrink=0.5, discrete=True)

# Add percentage labels
for container, senior in zip(gd.containers, total.index):
    counts = [p.get_height() for p in container]
    total_count = total[senior]
    percentages = [(count * 100) for count in counts]  # Scale to 100%

    # Apply the labels
    for bar, percentage in zip(container, percentages):
        gd.annotate(f'{percentage:.0f}%',
                    (bar.get_x() + bar.get_width() / 2, bar.get_y() + bar.get_height() / 2),
                    ha='center', va='center', fontsize=8, color='black', weight='bold')

plt.title("Churn By SeniorCitizen (Stacked Bar Chart)", fontsize=9)
plt.xlabel("Senior Citizen", fontsize=8)
plt.ylabel("Percentage (%)", fontsize=8)
plt.xticks(fontsize=8)

# Fix y-axis to show percentages from 0% to 100% in intervals of 10
plt.yticks(ticks=[i / 10 for i in range(0, 11)], labels=[f'{i * 10}' for i in range(0, 11)], fontsize=8)

# Move the legend to the top-right corner
plt.legend(title='Churn', labels=['Yes', 'No'], loc='upper right', bbox_to_anchor=(1.21, 1), fontsize=8)

plt.show()
```

```python
#comparitive a great percentage of people in senior citizen category have churned out.
```

```python
plt.figure(figsize=(12, 10))

columns = ['PhoneService', 'MultipleLines', 'InternetService',
           'OnlineSecurity', 'OnlineBackup', 'DeviceProtection',
           'TechSupport', 'StreamingTV']

# Loop through the columns and create a countplot for each
for i, col in enumerate(columns, 1):
    plt.subplot(4, 3, i)  # 4 rows, 2 columns
    sb.countplot(x=col, data=df, hue = "Churn")
    plt.title(f'Count plot of {col}')
# Move the legend to the top-right corner
    plt.legend(title='Churn', labels=['Yes', 'No'], loc='upper right', bbox_to_anchor=(1.25, 1), fontsize=8)
    plt.tight_layout()

plt.show()
```

```python
# Churn is higher for Fiber Optic users compared to DSL and those with no internet service.
# Customers without additional services like Online Security, Tech Support, and Device Protection show higher churn rates.
# Phone service users have lower churn, while those without phone service churn less overall.
# Streaming and backup services impact retention, but churn remains notable even among users of these services.
```
